```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

missing_values = ["Free"]
df = pd.read_csv("../../Datasets/Tweets.csv",na_values = missing_values)
#df = pd.read_csv("GraphicDesign.csv")
```

```python
from collections import Counter
data = df[["airline_sentiment","airline"]]

def summarise():
    """Summarises the number of positive, negative and neutral comments for each airline
    
    Args: None
    
    Returns: Returns a dataframe consisting of "Airline", "Comment sentiment type" and "Count"
    """
    table = []
    summarised_data = dict(Counter(zip(data["airline"],data["airline_sentiment"])))
    #print(summarised_data)
    for i in range(len(summarised_data)):
        table.append([list(summarised_data.keys())[i][0],list(summarised_data.keys())[i][1],summarised_data[list(summarised_data.keys())[i]]])
    df = pd.DataFrame(table, columns = ["Airline","Type","Count"])
    #ax = sns.scatterplot(x="Type", y="Count", data=df)
    return df.head(30)    
            
airline_data = summarise()
# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="Airline", y="Count", hue="Type",
              palette=["r", "c", "y"], data=airline_data)
```
